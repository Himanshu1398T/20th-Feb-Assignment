### Q1. Explain GET and POST methods.

#### Ans:
GET and POST are HTTP request methods used to retrieve and send data between a client and server in web applications.

The GET method is used to request a resource from a server. When a client sends a GET request, it asks the server to send back a specific resource, such as a webpage or a file. The request is sent in the URL of the request and can be easily bookmarked or shared with others. GET requests should only be used to retrieve data, and should not be used to modify data on the server.

The POST method is used to submit data to a server to create or update a resource. When a client sends a POST request, it includes data in the request body, which can include text, files, or other data types. This data is sent to the server, which can then use it to create or update a resource, such as adding a new record to a database.

### Q2. Why is request used in Flask?

#### Ans:
Request is a Python module used in Flask to handle incoming HTTP requests from clients, and to extract data from those requests.

In a Flask application, when a client sends an HTTP request to a server, Flask uses the Request module to create a request object that contains information about the request, such as the HTTP method used (GET, POST, etc.), any data or parameters sent with the request, and the URL of the requested resource.

The request object can be used in a Flask application to extract data from the request, such as form data, JSON data, or query parameters. This data can then be used to perform actions in the application, such as storing data in a database, generating a response to the client, or redirecting the client to a different page.

### Q3. Why is redirect() used in Flask?

#### Ans:
The redirect() function in Flask is used to redirect a client to a different URL.

In a Flask application, when a client sends an HTTP request to a server, the server generates a response that is sent back to the client. The response can include data such as HTML content, JSON data, or files. In some cases, it may be necessary to redirect the client to a different URL, such as when a user submits a form and should be redirected to a "Sucessfully Submitted Page" page.

### Q4. What are templates in Flask? Why is the render_template() function used?

#### Ans:
Templates in Flask are used to generate dynamic HTML content that can be sent back to clients in response to an HTTP request.

A template is a text file that contains a mixture of HTML and Python code, using special syntax that allows Python code to be executed and dynamic content to be generated.Flask uses a templating engine called Jinja2, which allows developers to easily create templates that can be rendered with dynamic data.
***
The render_template() function in Flask is used to render a template with dynamic data. When the function is called, Flask uses the Jinja2 templating engine to process the template file and replace any placeholders or expressions with data provided by the application. The resulting HTML content is then sent back to the client as a response to the HTTP request.

```python
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)


@app.route('/')
def home_page():
    return render_template('index.html')


@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        L1 = request.json['L1']
        if ops == 'Capital':
            r = L1.upper()
            result = f"The converted letter is {r}"

        if ops == 'Small':
            r = L1.lower()
            result = f"The converted letter is {r}"
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
